```python
df = pd.get_dummies(df, columns=['Furnishing', 'Status', 'Transaction', 'Type'], drop_first=True, dtype = int)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV 
```

```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=4)
```

```python
rf = RandomForestRegressor()
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr', lr), ('dt', dt), ('svr', svr), ('rf', rf)]
```

```python
for estimator in estimators:
    x = cross_val_score(estimator[1], X,y, cv = 15, scoring = 'r2')
    print(estimator[0], np.round(np.mean(x),2))
```

```python
parameters = {
    'n_estimators': [50, 100, 150, 200],          
    'max_depth': [None, 10, 20, 30, 40],                           
    'max_features': ['auto', 'sqrt', 'log2'],                         
    'criterion': ['squared_error', 'absolute_error', 'friedman_mse', 'poisson']                    
}
```

```python
grid = GridSearchCV(rf, param_grid=parameters, cv=5,verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
model = RandomForestRegressor(criterion='squared_error', max_depth=15, max_features='log2', n_estimators=150)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```
